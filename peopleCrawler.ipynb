```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import csv

# 初始化WebDriver
driver = webdriver.Chrome()

# 打開目標網站
driver.get('https://www.rootdata.com/people')

# 等待手動登錄
time.sleep(20)

def get_leading_figures():
    figures = []
    cards = driver.find_elements(By.CSS_SELECTOR, 'div.list_card.list_card_twitter')
    for card in cards:
        rank = card.find_element(By.CSS_SELECTOR, 'div.ranking').text
        name = card.find_element(By.CSS_SELECTOR, 'div.content h2').text
        role = card.find_element(By.CSS_SELECTOR, 'p.sub_color').text.strip()
        influence_index = card.find_element(By.CSS_SELECTOR, 'div.num_area span.text').text.strip()
        figures.append((rank, name, role, influence_index))
    return figures

def navigate_pages():
    page_number = 1
    all_figures = []
    while True:
        print(f"Currently on page {page_number}")
        all_figures.extend(get_leading_figures())
        try:
            # 確保下一頁按鈕在指定的div中
            twitter_wrapper = driver.find_element(By.ID, 'twitter_wraper')
            next_button = twitter_wrapper.find_element(By.CSS_SELECTOR, 'button.btn-next')
            
            if next_button and next_button.is_enabled():
                next_button.click()
                time.sleep(5)  # 等待下一頁加載
                page_number += 1
            else:
                print("No more pages or an error occurred.")
                break
        except Exception as e:
            print(f"An error occurred: {e}")
            break
    return all_figures

# 爬取所有頁面上的影響指數領導者數據
all_leading_figures = navigate_pages()

# 將數據保存到CSV文件
with open('leading_figures.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Rank', 'Name', 'Role', 'Influence Index'])
    writer.writerows(all_leading_figures)

# 關閉WebDriver
driver.quit()

print("爬取完畢")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

# 初始化 WebDriver
driver = webdriver.Chrome()

# 打開目標網站
driver.get('https://www.rootdata.com/people')

# 等待用戶手動登錄
time.sleep(20)  # 用戶有20秒時間來手動登錄

# 定義抓取函數
def get_universities():
    universities = []
    cards = driver.find_elements(By.CSS_SELECTOR, '.school_wraper .list_card')
    for card in cards:
        try:
            ranking = card.find_element(By.CSS_SELECTOR, '.ranking').text
            name = card.find_element(By.CSS_SELECTOR, 'h2').text
            link = card.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
            universities.append((ranking, name, link))
        except Exception as e:
            print(f"Error extracting data from card: {e}")
    return universities

# 定義翻頁函數
def navigate_pages():
    all_universities = []
    page_number = 1
    while True:
        print(f"Currently on page {page_number}")
        all_universities.extend(get_universities())
        try:
            # 查找並點擊下一頁按鈕
            wait = WebDriverWait(driver, 10)
            next_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.school_wraper .btn-next')))
            next_button.click()
            time.sleep(5)  # 等待下一頁加載
            page_number += 1
        except Exception as e:
            print(f"No more pages or an error occurred: {e}")
            break
    return all_universities

# 抓取所有頁面的數據
universities_data = navigate_pages()

# 關閉 WebDriver
driver.quit()

# 保存數據到 CSV 文件
with open('topRankUniversite.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Ranking", "University Name", "Link"])
    writer.writerows(universities_data)

print("Data has been written to topRankUniversite.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

# 初始化 WebDriver
driver = webdriver.Chrome()

# 打開目標網站
driver.get('https://www.rootdata.com/people')

# 等待用戶手動登錄
time.sleep(20)

def get_companies():
    companies = []
    try:
        cards = driver.find_elements(By.CSS_SELECTOR, '.member_wraper .list_card')
        for card in cards:
            try:
                ranking = card.find_element(By.CSS_SELECTOR, '.ranking').text
                name = card.find_element(By.CSS_SELECTOR, 'h2').text
                link = card.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
                companies.append([ranking, name, link])
            except Exception as e:
                print(f"Error in parsing card: {e}")
    except Exception as e:
        print(f"Error in finding cards: {e}")
    return companies

def navigate_pages():
    all_companies = []
    page_number = 1
    while True:
        print(f"Currently on page {page_number}")
        all_companies.extend(get_companies())
        try:
            # 檢查是否有下一頁按鈕，並確保點擊正確的按鈕
            wait = WebDriverWait(driver, 10)
            next_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.member_wraper .btn-next')))
            driver.execute_script("arguments[0].click();", next_button)
            time.sleep(5)  # 等待下一頁加載
            page_number += 1
        except Exception as e:
            print(f"No more pages or an error occurred: {e}")
            break
    return all_companies

# 抓取所有頁面的數據
companies_data = navigate_pages()

# 關閉 WebDriver
driver.quit()

# 保存數據到 CSV 文件
with open('topRankCompany.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Ranking", "Company Name", "Link"])
    writer.writerows(companies_data)

print("Data has been saved to topRankCompany.csv")
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# 初始化 WebDriver
driver = webdriver.Chrome()

# 打開 CSV 文件，讀取公司超連結
with open('topRankCompany.csv', mode='r') as file:
    reader = csv.DictReader(file)
    companies = list(reader)

def get_members(company_name, status):
    members = []
    container = driver.find_element(By.CSS_SELECTOR, 'div.container.association')
    member_cards = container.find_elements(By.CSS_SELECTOR, 'div.card_warper')
    for card in member_cards:
        try:
            img_url = card.find_element(By.CSS_SELECTOR, 'div.img_area img').get_attribute('src')
            name = card.find_element(By.CSS_SELECTOR, 'div.name').text
            current_company = card.find_element(By.CSS_SELECTOR, 'div.subtitle').text.strip()
            job_title = card.find_element(By.CSS_SELECTOR, 'div.introduce').text.strip()
            
            # 處理空白的當前公司和職位信息
            if not current_company:
                current_company = "N/A"
            if not job_title:
                job_title = "N/A"
            
            share_links = card.find_elements(By.CSS_SELECTOR, 'div.share a')
            x_link = ''
            linkedin_link = ''
            for link in share_links:
                href = link.get_attribute('href')
                if 'x.com' in href:
                    x_link = href
                elif 'linkedin.com' in href:
                    linkedin_link = href
            
            if status == "Current":
                current_company = company_name
            
            members.append({
                'company_name': company_name,
                'name': name,
                'status': status,
                'img_url': img_url,
                'current_company': current_company,
                'job_title': job_title,
                'x_link': x_link,
                'linkedin_link': linkedin_link
            })
        except Exception as e:
            print(f"Error in parsing card: {e}")
    return members

all_members = []

for company in companies:
    company_name = company['Company Name']
    company_link = company['Link']
    
    driver.get(company_link)
    time.sleep(5)
    
    # 獲取並處理前成員列表
    members = get_members(company_name, "Former")
    all_members.extend(members)
    
    # 嘗試切換到現任成員列表
    try:
        buttons = driver.find_elements(By.CSS_SELECTOR, 'div.tab button')
        for button in buttons:
            if "Current" in button.text:
                driver.execute_script("arguments[0].click();", button)
                time.sleep(5)
                members = get_members(company_name, "Current")
                all_members.extend(members)
                break
    except Exception as e:
        print(f"Error in switching to current members tab: {e}")

# 保存數據到 CSV 文件
with open('topRankCompanyFinal.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Company Name', 'Name', 'Status', 'Image URL', 'Current Company', 'Job Title', 'X Link', 'LinkedIn Link'])
    for member in all_members:
        writer.writerow([member['company_name'], member['name'], member['status'], member['img_url'], member['current_company'], member['job_title'], member['x_link'], member['linkedin_link']])

# 關閉 WebDriver
driver.quit()
```

```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# 初始化WebDriver
driver = webdriver.Chrome()

# 定義爬取數據的函數
def get_person_info():
    people = []
    container = driver.find_element(By.CSS_SELECTOR, 'div.container.association')
    cards = container.find_elements(By.CSS_SELECTOR, 'div.card_warper')

    for card in cards:
        try:
            image_url = card.find_element(By.CSS_SELECTOR, 'div.img_area img').get_attribute('src')
        except:
            image_url = "N/A"
        
        try:
            name = card.find_element(By.CSS_SELECTOR, 'div.name').text
        except:
            name = "N/A"
        
        try:
            company = card.find_element(By.CSS_SELECTOR, 'div.subtitle').text
        except:
            company = "N/A"
        
        try:
            position = card.find_element(By.CSS_SELECTOR, 'div.introduce').text
        except:
            position = "N/A"
        
        try:
            x_link = card.find_element(By.CSS_SELECTOR, 'a.links.twitter').get_attribute('href')
        except:
            x_link = "N/A"
        
        try:
            linkedin_link = card.find_element(By.CSS_SELECTOR, 'a.links.linkedin').get_attribute('href')
        except:
            linkedin_link = "N/A"
        
        people.append({
            'Name': name,
            'Image URL': image_url,
            'Current Company': company,
            'Job Title': position,
            'X Link': x_link,
            'LinkedIn Link': linkedin_link
        })
    
    return people

# 讀取topRankUniversity.csv
input_file = 'topRankUniversity.csv'
output_file = 'topRankUniversityPeople.csv'

with open(input_file, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    universities = list(reader)

# 爬取每個大學的成員信息並寫入CSV文件
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['University Name', 'Name', 'Image URL', 'Current Company', 'Job Title', 'X Link', 'LinkedIn Link']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for university in universities:
        university_name = university['University Name']
        link = university['Link']
        driver.get(link)
        time.sleep(5)  # 等待頁面加載

        people = get_person_info()
        
        for person in people:
            person['University Name'] = university_name
            writer.writerow(person)

# 關閉WebDriver
driver.quit()
```
